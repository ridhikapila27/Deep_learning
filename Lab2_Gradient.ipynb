```python
import numpy as np  #mathematical operations & arrays
import matplotlib.pyplot as plt #used for visualization
#pyplot is used to perform operations in a figure like plotting points,lines
def func(x): #def is used for user defined functions
    return 3*x**2 - 3*x + 4 #function given in question

def grad(x): #slope of curve is the derivative
    return 6*x - 3 #derivative of function

def gradient_descent(x0, learning_rate, num_iterations): #while creating a model,we define a cost function which should approach to zero 
#for minimum error.
    x = x0
    x_values = [x0]
    y_values = [func(x0)]
    for i in range(num_iterations):
        x = x - learning_rate*grad(x) #descent value after every iteration
        x_values.append(x) #append adds new item in the empty space 
        y_values.append(func(x)) 
    return x, x_values, y_values

x0 = 5 #initial value
learning_rate = 0.01
num_iterations = 100 #no. of epoch reduced if no of iterations reduced

min_x, x_values, y_values = gradient_descent(x0, learning_rate, num_iterations)

print("Minimum value: ", func(min_x))

x = np.linspace(-10, 10, 100)  #value of x ranges from -10 to 10 and number of iterations are fixed
y = func(x)
plt.plot(x, y, 'b-', label='Function') #displays data
plt.plot(x_values, y_values, 'r-', label='Gradient Descent')
plt.legend() #describes elements in the figure
plt.show() #displays figure
```
