```python
# define variables like input size, hidden unit, output size, batch size, and the learning rate.
n_input, n_out, batch_size, learning_rate = 10, 3, 100, 0.5
# randomly initialize the dummy input and the output target data
data_x = torch.randint(0,3,(batch_size,n_input)).float()
data_y = torch.randint(0,3,(batch_size,1)).float()
print(data_x.size())
print(data_y.size())
"""Building the Model"""
model = nn.Sequential(
nn.Linear(n_input,n_out),
nn.Sigmoid())
print(model)
"""define the loss function and the optimizer for gradient descent"""
loss_function = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
"""Training Loop"""
losses = []
for epoch in range(5000):
  # Forward propagation — compute the predicted y and calculate the current loss
  pred_y = model(data_x)
  loss = loss_function(pred_y, data_y)
  losses.append(loss.item()) #returns the value of the tensor as a standard python number
  # Backward propagation — after each epoch we set the gradients to zero before starting to do backpropagation
  model.zero_grad()
  loss.backward()
  # Gradient descent — Finally, we will update model parameters by calling optimizer.step() function
  optimizer.step()
"""Output"""
plt.plot(losses)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.title("Learning rate %f"%(learning_rate))
plt.show()
```
